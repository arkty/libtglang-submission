```python
import matplotlib.pyplot as plt
import numpy as np
import utils
players = list("!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~")

# Load all code files
lang_files = utils.load_lang_files()
```

```python
# Review one language PBI distribution
target_language = "TGLANG_LANGUAGE_C"
pbis = []
for file in lang_files[target_language]:
    pbis.append(utils.get_comb_phi_file(file.resolve(), players))
fig, ax = plt.subplots()
for idx, pbi in enumerate(pbis):
    ax.plot(pbi, label=f'{idx}')

fig.set_figwidth(12)    #  ширина Figure
fig.set_figheight(6)    #  высота Figure
x_axis = np.arange(len(players))
plt.ylabel("PBI")
plt.xlabel("Player")
plt.xticks(x_axis, players)
plt.show()
```

```python
# Review all languages PBI distribution

pbis = []
for l in lang_files:
    pbis.append(utils.get_comb_phi_file(lang_files[l][0].resolve(), players))
fig, ax = plt.subplots()
for idx, pbi in enumerate(pbis):
    ax.plot(pbi, label=f'{idx}')

fig.set_figwidth(12)
fig.set_figheight(6)
x_axis = np.arange(len(players))
plt.ylabel("PBI")
plt.xlabel("Player")
plt.xticks(x_axis, players)
plt.show()
```
